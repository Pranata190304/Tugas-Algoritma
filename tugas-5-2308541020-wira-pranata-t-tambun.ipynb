```python
def hitung_persegi (sisi) :
    luas = sisi ** 2
    keliling = 4 * sisi
    return (luas, keliling) 

sisi_persegi=4477889265179
hasil = hitung_persegi (sisi_persegi) 
luas, keliling = hasil

print("Luas persegi:", luas) 
print("keliling persegi:", keliling) 
```
