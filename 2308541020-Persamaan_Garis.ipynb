```python
import matplotlib.pyplot as plt
import numpy as np
# Tentukan rentang nilai untuk x
x_20 = np.linspace(-20, 20, 200)
# Tentukan persamaan garis lurus, misalnya y_20 = 2x_20 - 1
y_20 = 2 * x_20 - 1
# Membuat plot
plt.plot(x_20, y_20, label='y_20 = 2x_20 - 1')
# Menambahkan judul dan label sumbu
plt.title('Grafik Garis Lurus')
plt.xlabel('x_20')
plt.ylabel('y_20')
# Menambahkan grid
plt.grid(True)
# Menambahkan legenda
plt.legend()
# Menampilkan plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Definisikan dua persamaan garis lurus
# y = mx + c
def garis_lurus1(x):
    m1_20 = 2  # gradien garis pertama
    c1_20 = 1  # intercept garis pertama
    return m1_20 * x + c1_20
def garis_lurus2(x):
    m2_20 = -1  # gradien garis kedua
    c2_20 = 3  # intercept garis kedua
    return m2_20 * x + c2_20
# Buat array dari nilai x
x = np.linspace(-10, 10, 400)
# Hitung nilai y untuk setiap x
y1 = garis_lurus1(x)
y2 = garis_lurus2(x)
# Buat plot
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='y = 2x + 1')
plt.plot(x, y2, label='y = -x + 3')
# Tambahkan judul dan label
plt.title('Grafik Dua Persamaan Garis Lurus')
plt.xlabel('x')
plt.ylabel('y')
# Tambahkan grid
plt.grid(True)
# Tambahkan legenda
plt.legend()
# Tampilkan plot
plt.show()
```
